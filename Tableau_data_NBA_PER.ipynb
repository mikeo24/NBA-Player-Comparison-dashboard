```python
import pandas as pd 
import numpy as np
import requests
from bs4 import BeautifulSoup

# Install selenium because we are dealing with Javascript encoded website for players data

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from sklearn.preprocessing import StandardScaler

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
adv_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_advanced.html"
url = adv_stats_url.format(2023)
data = requests.get(url)
with open("NBA_playerdata_Tableau/per2023.html", "w+") as f:
    f.write(data.text)
```

```python
import time

year = 2023
url = adv_stats_url.format(year)

driver.get(url)
driver.execute_script("window.scrollTo(1,10000)") # Java script windowscroll - scroll the page down and grab the data 
time.sleep(2)

html= driver.page_source
```

```python
with open ("NBA_playerdata_Tableau/{}.html".format(year), "w+") as f: 
    f.write(html)
```

```python
df = []

year= 2023

with open ("NBA_playerdata_Tableau/{}.html".format(year)) as f:
    page = f.read()

soup = BeautifulSoup(page, "html.parser")
soup.find('tr', class_ = "thead").decompose()  
adv_table = soup.find(id="advanced_stats")
adv = pd.read_html(str(adv_table))[0]
adv["Year"] = year


df.append(adv)
```

```python
per = pd.concat(df)
```

```python
per.head()
```

```python
per.to_csv("advanced.csv")
```

```python
player_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"
url = player_stats_url.format(2023)
data = requests.get(url)
with open("NBA_playerdata_Tableau/2023.html", "w+") as f:
    f.write(data.text)
```

```python
import time

year = 2023
url = player_stats_url.format(year)

driver.get(url)
driver.execute_script("window.scrollTo(1,10000)") # Java script windowscroll - scroll the page down and grab the data 
time.sleep(2)

html= driver.page_source

with open ("NBA_playerdata_Tableau/{}.html".format(year), "w+") as f: 
    f.write(html)
```
